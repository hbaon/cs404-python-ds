```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Load CSV data from a URL
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'  # Titanic dataset example
df = pd.read_csv(url)

# Display the first few rows of the dataframe
print(df.head())
```

```python
# Step 2: Perform data transformation using numpy (normalize "Age" and "Fare")
# We will normalize the "Age" and "Fare" columns to a 0-1 scale using Min-Max normalization
age_min = df['Age'].min()
age_max = df['Age'].max()
fare_min = df['Fare'].min()
fare_max = df['Fare'].max()

df['Age_normalized'] = (df['Age'] - age_min) / (age_max - age_min)
df['Fare_normalized'] = (df['Fare'] - fare_min) / (fare_max - fare_min)

# Step 3: Visualize the normalized data using a scatter plot
plt.figure(figsize=(10, 6))

# Scatter plot of normalized "Age" vs "Fare"
plt.scatter(df['Age_normalized'], df['Fare_normalized'], alpha=0.5, color='blue')

# Add labels and title
plt.xlabel('Normalized Age')
plt.ylabel('Normalized Fare')
plt.title('Normalized Age vs Fare (Titanic Dataset)')

# Show the plot
plt.tight_layout()
plt.show()
```
